# Check Text Encoding

```python
#checks the encoding so the text can be read
import chardet
def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        raw_data = file.read()
        result = chardet.detect(raw_data)
        return result
#example
detect_encoding("C:/Users/Darren Zammit/Desktop/Thesis/datasets/heinrichi/A.txt")
```

# Create Collation

```python
from os import listdir
from os.path import isfile, join
from os import walk
import json
import re
from collatex import *

collation = Collation()

#tokenise the text so collatex can read it
def tokenize(input):
    words = re.split(r'\s+', input) # split on whitespace#r'.+\w|\W+$'
    tokens_by_word = [re.findall(r'.+$', word) for word in words]
    tokens = []
    for item in tokens_by_word:
        tokens.extend(item)
    token_list = [{"t": token} for token in tokens] # create dictionaries for each token
    return token_list

path="C:/Users/Darren Zammit/Desktop/Thesis/datasets/phareta fidei 10"
witnesses = next(walk(path), (None, None, []))[2]

corpus=[]
for i in witnesses:
    #put encoding from last part here
    j=open( f"{path}/{i}", encoding='utf-8' ).read()
    j=j.replace("\n", '').replace('  ', ' ').replace('-', "^")
    j_tokens=tokenize(j)
    i=i.replace(".txt","")
    witness_j={ "id": f"{i}", "tokens": j_tokens }
    #collation.add_plain_witness(f"{i}", j)
    corpus.append(witness_j)
    
input = {"witnesses": corpus }

alignment_table = collate(input, segmentation=False,near_match=True)
#this outputs ascii table 
print(alignment_table)
```

```python
import re
#make table with file name and contents into lists
x=re.split('\n',str(alignment_table))
x=x[1:len(x)]

table=[None] * len(x)

for i in range(len(x)):
    table[i]=x[i].split("|")[1:-1]
table=table[0:-1]
print(table)
```

```python
import pandas as pd
#turn into pandas data frame and save into a csv to manually edit it if need be
df=pd.DataFrame(table).drop(df.columns[0], axis=1)
df

#df.to_csv('parzival.csv', index=False)
#df=pd.DataFrame(table)
#df
#df=pd.read_csv('parzival.csv')
```

```python
#remove any trailing white spaces and the first column
d=df.loc[:, df.columns != 0]
for i in d.columns:
    d[i]=d[i].str.strip()
d
```

```python
import numpy as np
from scipy.stats import rankdata
from string import ascii_uppercase
#change words into pseudo-dna characters. identical words in a particular column get same the letter
#mismatches at same column get different letters
l=d
ind = rankdata(l, axis=0, method="dense") - 1
upl = np.vectorize(lambda x: ascii_uppercase[x])(ind)
out = pd.DataFrame(upl, columns=l.columns)
out

out2=out.mask(l=='-')
out2=out2.fillna('-')
out2
```

```python
#get the list of unique characters to specify this in nexus files
l=list(map(set,out.values))
from itertools import chain
ll=sorted(list(set(chain.from_iterable(l))))
characterlist=''.join(ll)
characterlist
```

```python
#create data blocks
out2['combined'] = out2[out2.columns[0:]].apply(lambda x: ''.join(x.dropna().astype(str)),axis=1)

datablocks=out2['combined'].tolist()
datablocks
```

```python
#number of characters in each sequence
nchar=len(out2.iloc[:, :-1].iloc[0])
nchar
```

```python
#create nexus file
file=open('phareta_fidei_10.nex','w')
file.write('#NEXUS\n')
file.write(f'BEGIN DATA;\n\tdimensions ntax = {len(exemplars)} nchar = {nchar};\n\tformat symbols = "{characterlist}" gap=- missing=? labels = left;\n\tmatrix')

for i in range(0,len(exemplars)):
    file.write(f"\n\t\t{exemplars[i]}\t{datablocks[i]}")
file.write(";\nEND;")
file.close()

# command to save a tree in paup*
#'savetrees file=treename.out format=altnex;'
#brlens allows you to save the branch lengths.
#savetrees file=treename.out brlens=yes format=altnex;
```
